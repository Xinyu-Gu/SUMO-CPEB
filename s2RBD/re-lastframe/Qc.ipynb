```python
d=[]
f=[]
index=np.loadtxt("filter1/cluster.index", dtype='int')
minf=np.loadtxt("filter1/miniF.dat", dtype='int')
print(index)
print(len(index))

for i in range(len(index)):
    d.append(c[index[i]])
    f.append(-minf[index[i]])
    
fig=plt.figure(figsize=(8,4.5))
ax = fig.add_subplot(111)
ax2 = ax.twinx()
lns1=ax.plot(d,color="b",label="E_AWSEM")
lns2=ax2.plot(f,color="r",label="-#minimal frustrations")
plt.xlim(0,14)
plt.xticks(np.arange(0,14,5))
ax.set_ylabel("E_AWSEM / (kcal/mol)", fontsize=20)
ax2.set_ylabel("-#minimal frustrations", fontsize=20)
ax.tick_params(axis='y', labelsize=20)
ax2.tick_params(axis='y', labelsize=20)

lns = lns1+lns2
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc=0,fontsize=16)
plt.tight_layout()
plt.savefig("qw_inter_energy.png")
```

```python
def Vbias(x):
    if x>15:
        V=1/2*0.1*(x-15)**2
    else:
        V=0
    return V
Vb=[]
X=np.arange(0,60,1)
for x in X:
    Vb.append(Vbias(x))

plt.figure(figsize=(5.5,4))
plt.plot(X,Vb,linewidth=3)
plt.tick_params(axis='x', labelsize=20)
plt.tick_params(axis='y', labelsize=20)
plt.ylabel("Vbias / (kcal/mol)", fontsize=22)
plt.xlabel("R / (A)", fontsize=22)
plt.xticks(np.arange(0,61,15))
plt.ylim(0,100)
plt.xlim(0,61)
plt.tight_layout()
plt.savefig("vbias.png")
```

```python
index=np.loadtxt("filter1/cluster.index", dtype='int')
minf=np.loadtxt("filter1/miniF.dat", dtype='int')
maxf=np.loadtxt("filter1/maxF.dat", dtype='int')
print(index)
print(len(index))
f=[]
g=[]

for i in range(len(index)):
    g.append(maxf[index[i]])
    f.append(-minf[index[i]])
    
fig=plt.figure(figsize=(8,4.5))
ax = fig.add_subplot(111)
ax2 = ax.twinx()
lns1=ax.plot(g,color="b",label="#high frustrations")
lns2=ax2.plot(f,color="r",label="-#minimal frustrations")
plt.xlim(0,14)
plt.xticks(np.arange(0,14,5))
ax.set_ylabel("#high frustrations", fontsize=20)
ax2.set_ylabel("-#minimal frustrations", fontsize=20)
ax.tick_params(axis='y', labelsize=20)
ax2.tick_params(axis='y', labelsize=20)

lns = lns1+lns2
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc=0,fontsize=16)
plt.tight_layout()
plt.savefig("frust.png")
```
